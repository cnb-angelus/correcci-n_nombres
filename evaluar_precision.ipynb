```python
def predictor_dummy(nombre_a_limpiar):
    if len(nombre_a_limpiar)<2:
        return nombre_a_limpiar
    if random.random() < 0.8:
        return nombre_a_limpiar
    else:
        LL = [l for l in nombre_a_limpiar]
        random.shuffle(LL)
        return "".join(LL)
```

```python
def precision_recall(true_col, pred_col):
    #Número de nombres verdaderos
    n_true = len(true_col)
    #Número de nombres predichos
    n_pred = len(pred_col)
    #Intersección de nombres verdaderos y predichos
    n_true_pred = len(set(true_col).intersection(set(pred_col)))
    #Precisión
    precision = n_true_pred/n_pred
    return precision, n_true_pred
```

```python
#Utilizar la columna transcrip para aplicar la función predictor_dummy
#Eliminar Nans
df=df.dropna(subset=['transcrip'])
df['predict']=df['transcrip'].apply(predictor_dummy)
#Crear una columna de 1 y 0 para saber si el nombre predicho es igual al nombre real
df["redultado"] = (df["predict"] == df["transcrip"]).astype(int)
```

```python
#Obtener precision y recall
precision,n_true_pred= precision_recall(df['transcrip'], df['predict'])
#Imprimir resultados
print('Precision: ', precision)
print('Número de nombres predichos: ', len(df['predict']))
print('Número de nombres predichos correctos: ', n_true_pred)
```
